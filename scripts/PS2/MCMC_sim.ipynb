```python
import numpy as np
import pandas as pd
```

```python
coords = np.loadtxt("../../data/init_crds_boxl_3.5-2.dat")
coords_df = pd.DataFrame(coords, columns=["x", "y", "z"])
coords_df
```

```python
def squared_distance(x1, x2, length):
    delta_x = x1 - x2
    delta_x -= length * np.round(delta_x/length)
    return delta_x
```

```python
distances = []
for i, particle_1 in enumerate(coords):
    for j in range(i + 1, len(coords)):
        particle_2 = coords[j]
        dist = np.linalg.norm(squared_distance(particle_1, particle_2, 3.5))
        distances.append(dist)
distances
```

```python
def lennard_jones_potential(dist, e, sigma):
    """
    Calculates the lennard jones potential for a given distance r_ij
    :param dist: The distance
    :param e: energy
    :param sigma: 'particle size'
    :return: Lennard Jones potential energy
    """
    return 4 * e * ((sigma / dist) ** 12 - (sigma / dist) ** 6)
```

```python
v = sum([lennard_jones_potential(dist, 0.25, 1) for dist in distances])
print(v)
```
